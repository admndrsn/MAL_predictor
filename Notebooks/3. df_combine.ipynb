### all_df -> This will be for the prediction / shows dataframe.

```python
all_df = pd.concat(map(pd.read_csv, glob.glob('[[]*[]] [[]ALL[]]*.csv')))

all_df = all_df.reset_index()
all_df['Genres'] = all_df["Genres"].apply(lambda x: ast.literal_eval(x))

#df['Genres'].fillna('NA', inplace = True)
flat_genre = [item for sublist in all_df["Genres"] for item in sublist]
set_genre = set(flat_genre)
unique_genre = list(set_genre)
#unique_genre = [item[1:-1] for item in unique_genre]
#unique_genre.remove('NA')
all_df = all_df.reindex(all_df.columns.tolist() + unique_genre, axis=1, fill_value=0)

for index, row in all_df.iterrows():
    for val in row.Genres:
        all_df.loc[index, val] = 1



all_df
```
